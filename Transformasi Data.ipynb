```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\MyWindows\AppData\Local\Programs\Python\Python37-32\DataMining\Data\data alumni.csv')
df
```

```python
df['ipk'] = df['ipk'].apply(str)
df['ipk'] = df['ipk'].str.replace(',','.').apply(float)

df
```

```python
df.dtypes
```

```python
for index, row in df.iterrows():
    if row['ipk'] >3.50:
        df.loc[index, 'ipk'] = 'Maksimum'
    elif row['ipk'] >= 2.75:
        df.loc[index, 'ipk'] = 'Rata - rata'
    else:
        df.loc[index, 'ipk'] = 'Minimum'
 
df
```

```python
data = df.dropna()
data
```

```python
df['TOEFL'] = df['TOEFL'].apply(str)
df['TOEFL'] = df['TOEFL'].str.replace(',','.').apply(float)

df
```

```python
for index, row in df.iterrows():
    if row['TOEFL'] >500:
        df.loc[index, 'TOEFL'] = 'T3'
    elif row['TOEFL'] >=450:
        df.loc[index, 'TOEFL'] = 'T2'
    else:
        df.loc[index, 'TOEFL'] = 'T1'
df[['NIM','ipk','TOEFL']]
```

```python
for index, row in df.iterrows():
    if row['Lama'] >9:
        df.loc[index, 'Lama'] = 'W4'
    elif row['Lama'] >= 6:
        df.loc[index, 'Lama'] = 'W3'
    elif row['Lama'] >= 3:
        df.loc[index, 'Lama'] = 'W2'
    else:
        df.loc[index, 'Lama'] = 'W1'
 
df[['NIM','ipk','TOEFL','Lama']]
```

```python
def_status = 'Mahasiswa'
df.status.fillna(def_status, inplace = True)
df
```
